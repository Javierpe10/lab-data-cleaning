#### 6. Define new dataframes for users and posts with the following selected columns:

```python
newdt_users = users_dt[["userId", "Reputation", "Views", "UpVotes", "DownVotes"]]
newdt_posts = posts_table[["postId", "Score", "userId", "ViewCount", "CommentCount"]]
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
mergedt = pd.merge(newdt_users, newdt_posts,
                   on="userId",
                   how="left"
                   )
mergedt.head()
```

```python
#Drop thecolumns that have more than half of rows with missing values. To scarce
#In this case we have one column (ViewCounts)
dt_null = mergedt.isnull().sum()[mergedt.isnull().sum()>32187]
mergedt_clean = mergedt.drop(columns=["ViewCount"])
#For the other columns with less than half of missing values we will fill them with the median
mergedt_final = mergedt_clean.fillna(mergedt.median())
mergedt_final.head()
```
